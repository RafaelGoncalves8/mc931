## Distancia

```python
from math import cos, asin, sqrt
def distance(lat1, lon1, lat2, lon2):
    #p = 0.017453292519943295     #Pi/180
    p = 1
    a = 0.5 - cos((lat2 - lat1) * p)/2 + cos(lat1 * p) * cos(lat2 * p) * (1 - cos((lon2 - lon1) * p)) / 2
    return 12742 * asin(sqrt(a)) #2*R*asin...
```

```python
LAT = df.at[3375, "lat"]
LON = df.at[3375, "lon"]
```

```python
for i, e in enumerate(df.cidade):
    df.at[i, "dist"] = distance(df.lat[i], df.lon[i], LAT, LON)
```

## Planilha ano-curso

```python
cursos = []

for e in data.nome_curso:
    if e not in cursos:
        cursos.append(e)
```

```python
ano_curso = pd.DataFrame(np.nan*np.ones((20,72)), index=range(1998,2018), columns=cursos)
ano_curso.head()
```

```python
def filter_by_year(ano):
    return df[df['ano'] == ano]

def filter_by_year_and_uf(ano, uf):
    fy = filter_by_year(ano)
    return fy[fy['UF'] == uf]

def filter_by_year_course(ano, curso):
    fy = filter_by_year(ano)
    return fy[fy['nome_curso'] == curso]

def filter_by_year_uf_course(ano, uf, curso):
    fyu = filter_by_year_and_uf(ano, uf)
    return fyu[fyu['nome_curso'] == curso]

def porcentagem_fora_SP_ano(ano):
    return (1 - sum(filter_by_year_and_uf(ano, 'SP')['numero']) / sum(filter_by_year(ano)['numero']))

def percentage_nSP(ano, curso):
    return (1 - sum(filter_by_year_uf_course(ano, 'SP', curso)['numero']) / 1)#sum(filter_by_year_course(ano, curso)['numero']))

```

```python
ano_curso_SPn = ano_curso.copy()
```

```python
for ano in ano_curso.index:
    data_ano = data[data["ano"] == ano]
    tot = sum(int(e) for e in data_ano.numero)
    SP = data_ano[data_ano["UF"] == "SP"]
    SPn = data_ano[data_ano["UF"] != "SP"]
    for curso in cursos:
        ano_curso.at[ano, curso] = sum([int(e) for e in SP[SP["nome_curso"] == curso].numero])
        ano_curso_SPn.at[ano, curso] = sum([int(e) for e in SPn[SPn["nome_curso"] == curso].numero])
```
